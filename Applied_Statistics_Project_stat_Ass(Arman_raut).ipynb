```python
# Identify variables that could follow specific probability distributions:
# Binomial distribution:
binomial_variables = ['Complain', 'Response']

# Normal distribution:
normal_variables = ['Year_Birth', 'Income', 'MntFishProducts']


# Calculate probabilities and expected values for binomial variables:
# Probability of a customer complaining:
prob_complain = df['Complain'].mean()
print(f"Probability of a customer complaining: {prob_complain}")

# Probability of a customer accepting the offer:
prob_response = df['Response'].mean()
print(f"Probability of a customer accepting the offer: {prob_response}")


# Calculate probabilities and expected values for normal variables:
# Expected value of customer age:
expected_age = df['Year_Birth'].mean()
print(f"Expected value of customer age: {expected_age}")

# Standard deviation of customer income:
std_income = df['Income'].std()
print(f"Standard deviation of customer income: {std_income}")
```

## Task 4: Insights and Customer Segmentation üìà

- **Explore relationships between customer characteristics and spending habits. Segment customers based on their behaviors and characteristics.**

```python
# Analyze spending habits by age group:
age_groups = df.groupby('Year_Birth')['MntWines'].mean()
plt.bar(age_groups.index, age_groups.values)
plt.xlabel('Age Group')
plt.ylabel('Average Spending on Wines')
plt.title('Spending on Wines by Age Group')
plt.show()
```

```python
# Segment customers based on spending behaviours:
df['Spending_Segment'] = pd.cut(df['MntWines'], bins=3, labels=['Low', 'Medium', 'High'])

# Analyze customer segments:
segment_analysis = df.groupby('Spending_Segment').agg({'Response': 'mean', 'Income': 'mean', 'Kidhome': 'mean'})
print(segment_analysis)
```

**Deliverables**:

- **Key insights regarding relationships between variables and distinct customer segments based on behaviors.**
